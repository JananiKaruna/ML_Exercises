```python
df['YearsCodeStr'] = df.YearsCode.astype('str')
```

```python
def change_str_values(value):
    str_value = value.strip()
    if str_value == 'nan':
        return '0'
    elif str_value == 'More than 50 years':
        return '50'
    elif str_value == 'Less than 1 year':
        return '0'
    else:
        return str_value
        
```

```python
df['YearsCodeStr2'] = df['YearsCodeStr'].apply(change_str_values)#, axis=1)
```
